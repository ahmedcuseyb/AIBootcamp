### 1- Manual Encoding Mapping

```python
# Clean strings

df["Neighborhood"] = df["Neighborhood"].astype(str).str.strip()
df["Condition"] = df["Condition"].astype(str).str.strip()

# Manual mapping dictionaries
neighborhood_map = {
    "Low-Cost": 0,
    "Outskirts": 1,
    "Middle Class": 2,
    "Expensive Area": 3
}

condition_map = {
    "Poor": 0,
    "Fair": 1,
    "Good": 2,
    "Excellent": 3
}

# Apply mapping
df["Neighborhood"] = df["Neighborhood"].map(neighborhood_map)
df["Condition"] = df["Condition"].map(condition_map)
```
